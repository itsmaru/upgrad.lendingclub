```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("loan.csv",low_memory=False)
```

### 1. Preprocessing

#### 1.1 Handling Missing Value

Check the distribution of missing value on the data

```python
# Function for plotting displot
def plot_missing_value(df):
    plt.figure(figsize=(30,30))
    sns.displot(
        data=df.isna().melt(value_name="missing"),
        y="variable",
        hue="missing",
        multiple="fill",
        # aspect=1.25,
        # vertical=True
        height=15
    )
    plt.show()
```

```python
plot_missing_value(df)
```

Seems theres a lot of missing data in this data set, some have 100% missing value counts, it's better to drop all the columns that have 100% missing values.

```python
# drop all the columns which have more than 95% of missing values
df.dropna(thresh=df.shape[0]*0.95,how='all',axis=1,inplace=True)
```

```python
df
```

Lets check again the distribution of missing values

```python
plot_missing_value(df)
```

looks great, but theres some missing value, which will be filled by using the most frequent value from that columns

```python
# Locate the columns which contains missing value
columns_na = df.columns[df.isna().any()].tolist()
columns_na
```

```python
# and fill the missing value using the ost frequent value
for column in columns_na:
    df[column] = df[column].fillna(df[column].value_counts().index[0])
```

```python
plot_missing_value(df)
```

Looks great!

#### 1.2 Handling Constant Value Columns
theres no benefit of a variable/columns if theres only a single constant value in it, hence it will be removed

```python
# Get the columns/variable which only consist of 1 distinct value
columns_constant = df.columns[df.nunique() <= 1]
columns_constant
```

```python
df.drop(columns_constant,inplace=True,axis=1)
```

```python
df
```

### 1.3 Handle Skewness in data
skewness in a columns/variable, isn't great to be feed to predictive modeling, it might cause bias in the future predictions, its best to handle it so that it has equal distribution

```python
# Check skewness of more than 10
numeric_columns = df.select_dtypes(exclude="object").columns
skew_columns = df[numeric_columns].loc[:,(df[numeric_columns].skew() > 10).to_list()].columns
df[skew_columns].skew()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(18, 10))
for index, column in enumerate(skew_columns):
    axes[index].set_title(column)
    sns.histplot(data=df[column],ax=axes[index])
```

```python
# Handle the skew data by sqrt the columns
df[skew_columns] = np.sqrt(df[skew_columns])
df[skew_columns].skew()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(18, 10))
fig.tight_layout()
for index, column in enumerate(skew_columns):
    sns.histplot(data=df[column],ax=axes[index])
```

#### 1.4 Handle Categorical Value
categorical value cant be read to machine learning model thus we need to convert every categorical data using ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
categorical_columns = df.select_dtypes(include="object").columns
df[categorical_columns] = oe.fit_transform(df[categorical_columns])
```

```python
df
```

### 2. Modelling

#### 2.1 Split the dataset
the dataset will be split to 80% training and 20% test set

```python
y = df['loan_status']
X = df.drop('loan_status',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2) 
```

#### 2.2 Modeling using Decision Tree

```python
decision_tree = DecisionTreeClassifier()
decision_tree = decision_tree.fit(X_train, y_train)
```

```python
predictions = decision_tree.predict(X_test)
```

```python
cr = pd.DataFrame(classification_report(predictions,y_test, digits=2,output_dict=True)).T
cr['support'] = cr.support.apply(int)
cr.style.background_gradient(cmap='viridis',
                             subset=pd.IndexSlice['0':'9', :'f1-score'])
```
